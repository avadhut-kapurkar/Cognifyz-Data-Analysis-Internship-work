```python
import pandas as pd
import folium
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df[['Latitude','Longitude']].head()
```

```python
map_restaurants = folium.Map(
    location=[df['Latitude'].mean(),
              df['Longitude'].mean()],
    zoom_start=5
)
map_restaurants
```

```python
for index, row in df.head(100).iterrows():
    folium.Marker(
        [row['Latitude'], row['Longitude']]
    ).add_to(map_restaurants)

map_restaurants
```

```python
map_restaurants.save("restaurant_map.html")
```

Conclusion ==

Restaurant locations were plotted using latitude and longitude coordinates

The map shows clusters of restaurants concentrated in major urban areas, indicating higher restaurant density in populated cities
