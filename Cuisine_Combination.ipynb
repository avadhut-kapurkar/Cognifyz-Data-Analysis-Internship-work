```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df['Cuisines'].head()
```

```python
df['Cuisines'].value_counts().head(20)
```

```python
top_combo = df['Cuisines'].value_counts().head(20)
top_combo.plot(kind='bar')
plt.title("Top 10 Cuisine Combinations")
plt.xlabel("Cuisine Combination")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(20)
```

```python
df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(1)
```

Conclusion ==

The most common cuisine combinations were identified from the dataset

Some cuisine combinations consistently received higher ratings than others, indicating customer preference
