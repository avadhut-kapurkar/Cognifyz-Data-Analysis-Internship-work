```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
pd.crosstab(
    df['Price range'],
    df['Has Online delivery']
)
```

```python
pd.crosstab(
    df['Price range'],
    df['Has Online delivery']
).plot(kind='bar')

plt.title("Price Range vs Online Delivery")

plt.show()
```

```python
pd.crosstab(
    df['Price range'],
    df['Has Table booking']
)
```

```python
pd.crosstab(
    df['Price range'],
    df['Has Table booking']
).plot(kind='bar')

plt.title("Price Range vs Table Booking")

plt.show()
```

```python
online_delivery_percentage = pd.crosstab(
    df['Price range'],
    df['Has Online delivery'],
    normalize='index'
) * 200

print(online_delivery_percentage)
```

```python
table_booking_percentage = pd.crosstab(
    df['Price range'],
    df['Has Table booking'],
    normalize='index'
) * 100

print(table_booking_percentage)
```

Conclusion ==

The relationship between price range, online delivery, and table booking was analyzed

Higher-priced restaurants tend to offer table booking services more frequently

Online delivery availability varies across different price categories
