```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df.nlargest(1, 'votes')[['restaurant Name','Votes']]
```

```python
df.nsmallest(1, 'Votes')[['Restaurant Name','Votes']]
```

```python
plt.scatter(df['votes'], df['aggregate rating'])
plt.title("votes vs Ratings")
plt.xlabel("votes")
plt.ylabel("aggregate Rating")
plt.show()
```

```python
df[['Votes', 'Aggregate rating']].corr()
```

Conclusion ==

The restaurants with the highest and lowest votes were identified

A correlation analysis was performed between votes and aggregate ratings to understand whether popular restaurants tend to receive higher ratings
