```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df.head()
```

```python
df['City'].value_counts().head(20)
```

```python
top_cities = df['City'].value_counts().head(20)

top_cities.plot(kind='bar')

plt.title("top 20 cities with restaurants")

plt.show()
```

```python
df.groupby('City')['Aggregate rating'].mean()
```

```python
df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(1)
```

Conclusion ==
new Delhi  has the highest number of restaurants in the dataset.
