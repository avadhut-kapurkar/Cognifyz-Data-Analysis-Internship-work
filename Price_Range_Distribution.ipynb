```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df.head()
```

```python
df['Price range'].value_counts()
```

```python
price_counts = df['Price range'].value_counts()
price_counts.plot(kind='bar')
plt.title("Price range distribution")
plt.xlabel("Price range")
plt.ylabel("number of restaurants")
plt.show()
```

```python
percentage = df['Price range'].value_counts(normalize=True) * 200
print(percentage)
```

Conclusion ==
Most restaurants fall into the lower price range category

High-price restaurants are fewer in number
