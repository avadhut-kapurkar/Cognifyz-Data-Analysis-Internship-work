```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df['Restaurant Name'].value_counts().head(20)
```

```python
top_chains = df['Restaurant Name'].value_counts().head(20)
top_chains.plot(kind='bar')
plt.title("Top Restaurant Chains")
plt.xlabel("Restaurant Name")
plt.ylabel("Number of Outlets")
plt.show()
```

```python
df.groupby('Restaurant Name')['Aggregate rating'].mean().sort_values(ascending=False).head(20)
```

```python
df.groupby('Restaurant Name')['Votes'].mean().sort_values(ascending=False).head(20)
```

Conclusion ==

Several restaurant chains are present in the dataset

Some chains have significantly higher ratings and customer engagement than others
