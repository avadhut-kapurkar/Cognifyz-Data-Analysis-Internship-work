```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df.head()
```

```python
df['Has Online delivery'].value_counts()
```

```python
delivery_percentage = df['Has Online delivery'].value_counts(normalize=True) * 100
print(delivery_percentage)
```

```python
delivery_counts = df['Has Online delivery'].value_counts()
delivery_counts.plot(kind='bar')
plt.title("Online Delivery Availability")
plt.xlabel("Online Delivery")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
df.groupby('Has Online delivery')['Aggregate rating'].mean()
```

Conclusion:
A smaller percentage of restaurants offer online delivery

Restaurants with online delivery tend to have better ratings
